# Brownian motion

In this notebook (side project/good to know topic), the goal is to understand the code behind Brownian motion as a stochastic process. This is meant to develop intuition behind random walk.

Wiener process
A 
real valued continous-time stochastic process with stationary independent increments and find frequent applications in a wide range of fields cvering pure and applied mathematics, economics, quantitative finance, evolutionary biology and physics. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gen_random_walk() : Generates motion from the Random Walk process
# gen_normal() : Generates motion by drawing from the Normal Distribution
# stock_price() : Models a stock price using so called 'Geometric Brownian Motion'ArithmeticError

class Brownian():
    """
    A Brownian motion class constructor
    """
    def __init__(self, x0=0):

        assert (type(x0)==float or type(x0)==int or x0 is None), "Expect a float"

        self.x0 = float(x0)

    def gen_random_walk(self, n_step=100):
        if n_step <30:
            print("Warnning! Small number")
        
        w = np.ones(n_step)*self.x0

        for i in range(1, n_step):

            yi = np.random.choice([1,-1])

            w[i] = w[i-1]+(yi/np.sqrt(n_step))

        return w
    
    
    def gen_normal(self, n_step=100):
        if n_step <30:
            print("Warnning! Small number")
        
        w = np.ones(n_step)*self.x0

        for i in range(1, n_step):

            yi = np.random.normal()

            w[i] = w[i-1]+(yi/np.sqrt(n_step))

        return w
    
    def stock_price(
                    self, 
                    s0=100,
                    mu=0.2,
                    sigma=0.68,
                    deltaT=52,
                    dt=0.1
                    ):
        n_step = int(deltaT/dt)
        time_vector = np.linsepace(0, deltaT, num=n_step)
        stock_var = (mu-(sigma**2/2)*time_vector)
        self.x0=0
        weiner_process = sigma*self.gen_normal(n_step)
        s = s0(np.exp(stock_var+weiner_process))
        return s
    
    
        
```

```python
b = Brownian()
```

```python
for i in range(4):
    plt.plot(b.gen_random_walk(10000))
    plt.show()
```
