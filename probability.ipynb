## Bernoulli distribution
The probability distribution od a random variable which takes the value 1 with probability p and the value 0 with probability q=1-p. 

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy. stats import bernoulli
```

```python
# Generate random variables
bernoulli. rvs(p=0.5, size=10)
```

```python
bernoulli.rvs(p=0.5, size=20)
```

```python
N_trials = [10, 25, 100, 250, 1000, 10000]
pr=0.5 #coin toss probability
av = [] # to store average of the random variates

# Generate 10 variates every time and take the average
for i in N_trials:
    for n in range(1,i+1):
        av.append(np.mean(bernoulli.rvs(p=pr,size=10)))
    if(i==10):
        plt.title("Distribution with {} trials of 10 coin tosses".format(i))
        plt.hist(av, bins=10)
        plt.xlim(0.0, 1.0)
        plt.show()
    else:
        plt.title("Distribution with {} trials of 10 coin tosses".format(i))
        plt.hist(av, bins=10)
        plt.xlim(0.0, 1.0)
        plt.show()

# From the observation, the distribution centres around 0.5
```

```python
pr = 0.5 # fair coin toss probability
mean, var, skew, kurt = bernoulli.stats(p=pr, moments='mvsk')
print(mean)
print(kurt)
# Biased coin 
pr = 0.7
mean, var, skew, kurt = bernoulli.stats(p=pr, moments='mvsk')
print(mean)
print(kurt)
```

Probability mass function (pmf) and cumulative distribution function (cdf)

```python
rv = bernoulli(0.6)
x = 0
print("Probability mass function for {}:{}".format(x, rv.pmf(x)))
x = 0.5
print("Probability mass function for {}:{}".format(x, rv.pmf(x)))
x = 1.0
print("Probability mass function for {}:{}".format(x, rv.pmf(x)))
x = 1.2
print("Probability mass function for {}:{}".format(x, rv.pmf(x)))
```

```python
print("CDF for x < 0:", rv.cdf(-2))
print("CDF for 0 < x <1", rv.cdf(0.75))
print("CDF for x >1:", rv.cdf(2))
```

# Binomial Distribution

The binomial distribution with parameter n and p is the discrete probability distribution of successes in a seuence of n independent experiments, each asking a yes-no qustion, and each with its own boolean-valued outcome: a random variable containing single bit of information: success/yes/true/one (with probability p) or failure/no/false/zero (with probability q=1-p).

The binomial distribution is frequently useed to model the number of successes in a sample of size n drawn with replacement from a population size N. If the sampling is carried out with replacement, the draws are not independent and so resulting distribution is a hypergeometric distribution, not a binomial one.

```python
from scipy.stats import binom

# Generate random variables
# A coin is flipped 8 times, each with probability of success (1) of 0.75
# experiement is repeated 10 times
k = binom.rvs(8, 0.75, size=10)
print('Number of success for each trial:', k)
print('Average of the success:', np.mean(k))
```

```python
# Visualizing the probability mass function

n = 40
pr = 0.5
rv = binom(n,pr)
x = np.arange(0,41,1)
pmf1 = rv.pmf(x)

n = 40
pr = 0.75
rv = binom(n,pr)
x = np.arange(0,41,1)
pmf2 = rv.pmf(x)

n = 40
pr = 0.35
rv = binom(n,pr)
x = np.arange(0,41,1)
pmf3 = rv.pmf(x)


plt.figure(figsize=(12,6))
plt.title('Probability mass function')
plt.scatter(x,pmf1)
plt.scatter(x,pmf2)
plt.scatter(x,pmf3,c='k')

plt.legend(["$n=40, p=0.5$","$n=40, p=0.3$","$n=50, p=0.6$"])
plt.grid(True)
plt.show()
```

# Poisson Distribution

A discrete probability distribution that expresses the probability of a given number of events occuring in a fixed interval of time or space if these events occur with a known constant rate and independently of the time since the last event.
Example:
- Number of patients arriving in an emergency room between 10 and 11 pm

Poisson distribution is a limiting case of a Binomial Distribution where the number of trials is sufficiently bigger than the number of successes one is asking about. i.e. n>>1>>p** 

```python
from scipy.stats import poisson
```

 ### Display probability mass function (pmf)
 An event can occur 0,1,2, ... times in an interval. The average number of events in an interval is designated λ. This is the event rate, also called the rate parameter. The probability of observing k events.

```python
la = 0.5 # la represents the average rate at which evebts occur
rv = poisson(la) # this line creates a Poisson random variable 'rv'
x = np.arange(0,11,1) #  It starts from 0, ends at 10 (exclusive), and increments by 1. 
pmf1 = rv.pmf(x) # This line computes the probability mass function (PMF) values for the Poisson distribution at each point in the array x. 

plt.figure(figsize=(9,6))
plt.scatter(x,pmf1,s=100)
plt.show()
```
