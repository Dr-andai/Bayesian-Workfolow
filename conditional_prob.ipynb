# Newton-Pepys Problem: An Insight inro Probability

In this section of code, we will be seeking inspiration of coding more on Probability through the Newton-Pepys Problem. The code is inspired from this link 
https://medium.datadriveninvestor.com/the-newton-pepys-problem-an-insight-into-probability-through-mathematics-and-python-simulations-e3a2d6c51080

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initializing lists to store probabilities
p_fair = []
p_biased_6 = []
p_biased_1 = []

n_matches_fair = []
n_matches_biased_6 = []
n_matches_biased_1 = []

# Loop Over different scenarios
for situation_num in range (1,20):

    # Calculate the number of dice and the number of matches
    n_dice = 6*situation_num
    n_matches = 1*situation_num
    n_trials = 10**6
    sequence = [1,2,3,4,5,6]


    # Generate dice rolls for each scenario
    rolls_fair = np.random.randint(1,7, size=n_dice*n_trials).reshape(n_trials,n_dice)
    rolls_biased_6 = np.random.choice(sequence, (n_trials, n_dice),p = [0.1,0.15,0.15,0.15,0.20,0.25])
    rolls_biased_1 = np.random.choice(sequence, (n_trials, n_dice),p = [0.25,0.2,0.15,0.15,0.15,0.1])

    # Calculate the number of sixes for fair dice
    sixes_fair = rolls_fair == 6

    # Calculate the number of wins for fair dice
    wins_fair = np.sum(sixes_fair, axis=1) >= n_matches

    # Calculate the number of wins fair dice
    n_wins_fair = np.sum(wins_fair)

    # Calcualte the number of wins for fair dice
    p_success_fair = n_wins_fair/n_trials

    # Append the success probability and number of matches to the respective lists
    p_fair.append(p_success_fair)
    n_matches_fair.append(n_matches)

    ## Repeat the above process for dice biased towards 6
    sixes_biased_6 = rolls_biased_6 == 6
    wins_biased_6 = np.sum(sixes_biased_6, axis=1) >= n_matches
    n_wins_biased_6 = np.sum(wins_biased_6)
    p_success_biased_6 = n_wins_biased_6/n_trials
    p_biased_6.append(p_success_biased_6)
    n_matches_biased_6.append(n_matches)

    ## Repeat the above steps for dice biased towards 1
    sixes_biased_1 = rolls_biased_1 == 6
    wins_biased_1 = np.sum(sixes_biased_1, axis=1) >= n_matches
    n_wins_biased_1 = np.sum(wins_biased_1)
    p_success_biased_1 = n_wins_biased_1/n_trials
    p_biased_1.append(p_success_biased_1)
    n_matches_biased_1.append(n_matches)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot (n_matches_fair, p_fair, linewidth=2.0)
ax.plot (n_matches_biased_6, p_biased_6, linewidth=2.0)
ax.plot (n_matches_biased_1, p_biased_1, linewidth=2.0)
```
